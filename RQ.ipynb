## Tariff models and where to define them

For RQ1 we want to compare **multiple tariff structures** (e.g., flat, TOU, dynamic/RTP proxy)
on the **same set of household load profiles**.

We structure this in two layers:

1. **Tariff pricing functions** (one per tariff)  
   Each takes timestamps (and optionally metadata) and returns a per-interval price in \$/kWh.  

2. **Generic bill calculator**  
   Given a timeseries DataFrame with `bldg_id`, `timestamp`, and `kwh_total`, plus a pricing
   function, compute:
   - `annual_kwh`
   - `annual_cost_<tariff_name>` per building

This design lets us swap in different pricing datasets or tariff logic **without changing the rest of the notebook**, and now scales efficiently to thousands of buildings.

```python
# Compute bills for all buildings using chunked processing
USE_FULL_CHUNKED = True
CHUNK_SIZE = 100  # Adjust based on available RAM (50-200 is reasonable)

if not USE_FULL_CHUNKED:
    # Sample-based version (fast, lightweight)
    print("Computing bills on SAMPLE only...")
    bills_sample = apply_all_tariffs(ts_sample, meta_df=meta_sample)
    print("bills_sample shape:", bills_sample.shape)
    display(bills_sample.head())

    if meta_sample is not None:
        bills_with_meta = meta_sample.merge(bills_sample, on="bldg_id", how="inner")
        print("bills_with_meta shape:", bills_with_meta.shape)
    else:
        bills_with_meta = None
        print("No metadata available; cannot attach attributes.")

else:
    # Full dataset via chunked processing (verbose=False for clean output)
    print(f"Computing bills on FULL dataset ({len(ts_index)} buildings) in chunks of {CHUNK_SIZE}...")

    bills_all = compute_bills_chunked(
        ts_index=ts_index,
        elec_cols=elec_cols,
        meta_df=meta_sample,
        chunk_size=CHUNK_SIZE,
        verbose=False,  # Suppress detailed progress messages
    )
    print("bills_all shape:", bills_all.shape)
    display(bills_all.head())

    # Join with metadata to get demographic attributes
    if meta_sample is not None:
        bills_with_meta = meta_sample.merge(bills_all, on="bldg_id", how="inner")
        print("bills_with_meta shape:", bills_with_meta.shape)
    else:
        bills_with_meta = None
        print("No metadata available; cannot attach attributes.")
    
    print("✓ Bill computation complete")
```

## RQ2: Distributional Fairness

**Analysis Goals:**
1. Examine how bills and energy burdens vary across demographic groups
2. Identify vulnerable populations facing disproportionate costs
3. Assess whether alternative tariffs improve or worsen equity

**Demographic Dimensions:**
- **Income**: Federal Poverty Level categories (0-100%, 100-150%, 150-200%, 200-300%, 300-400%, 400%+)
- **Tenure**: Homeowners vs. Renters
- **Geography**: Urban (principal city) vs. Suburban vs. Rural (non-metro)

**Key Metric**: PIU (Percentage of Income spent on Utilities) = (Annual Bill / Annual Income) × 100
- **3% threshold**: Moderate energy burden
- **6% threshold**: High energy burden (indicates energy insecurity)

### Cost per kWh
Here we examine how cost per kWh under each tariff (flat, TOU, dynamic) vary across household income levels as measured by Federal Poverty Level.  
For each poverty level category, we compare:
- **Boxplots** showing the distribution of cost per kWh under each of the three tariff structures
- **Summary statistics** (count, mean, median, std, min, max) for each tariff × poverty level combination

This analysis helps us understand whether different tariff structures create disparate impacts across income groups.

```python
# Compute cost per kWh
if "annual_kwh" in df_inc.columns:
    # Create a temporary copy with per_kwh columns just for analysis
    # This avoids SettingWithCopyWarning and doesn't modify bills_with_meta
    df_per_kwh = df_inc.copy()
    
    for col in bill_cols:
        df_per_kwh[f"{col}_per_kwh"] = df_per_kwh[col] / df_per_kwh["annual_kwh"]
    
    # Now you can do the same summaries / boxplots but for $/kWh
    for col in bill_cols:
        per_kwh_col = f"{col}_per_kwh"
        
        # Calculate y-axis limits individually for this tariff
        # Lower bound: use 2nd percentile or 0, whichever is lower (handles negative outliers)
        y_min = min(0, df_per_kwh[per_kwh_col].quantile(0.02))
        # Upper bound: 98th percentile to exclude extreme outliers
        y_max = df_per_kwh[per_kwh_col].quantile(0.98)
        
        #print(f"\n{col}: Setting y-axis limits to [{y_min:.4f}, {y_max:.4f}] $/kWh")
        
        plt.figure(figsize=(6, 4))
        df_per_kwh.boxplot(
            column=per_kwh_col,
            by="in.federal_poverty_level",
            grid=False,
        )
        plt.title(f"{col} per kWh by federal poverty level")
        plt.suptitle("")  # remove default pandas title
        plt.xlabel("Federal poverty level")
        plt.ylabel("Cost per kWh ($/kWh)")
        plt.ylim(y_min, y_max)  # Set y-axis to show 2nd to 98th percentile clearly
        plt.tight_layout()
        plt.show()

    # Numeric summary
    summary_per_kwh = (
        df_per_kwh.groupby("in.federal_poverty_level")[
            [f"{col}_per_kwh" for col in bill_cols]
        ]
        .agg(["mean", "std", "min", "max"])
    )
    summary_per_kwh.columns = [
        f"{tariff}_per_kwh_{stat}" for tariff, stat in summary_per_kwh.columns
    ]
    print("\nSummary of cost per kWh by income quantile:")
    display(summary_per_kwh)
    
    # Report extreme outliers for transparency
    print("\n=== Outlier Analysis: Cost per kWh ===")
    for col in bill_cols:
        per_kwh_col = f"{col}_per_kwh"
        y_min = min(0, df_per_kwh[per_kwh_col].quantile(0.02))
        y_max = df_per_kwh[per_kwh_col].quantile(0.98)
        min_val = df_per_kwh[per_kwh_col].min()
        max_val = df_per_kwh[per_kwh_col].max()
        p95 = df_per_kwh[per_kwh_col].quantile(0.95)
        p98 = y_max
        p99 = df_per_kwh[per_kwh_col].quantile(0.99)
        n_extreme_high = (df_per_kwh[per_kwh_col] > y_max).sum()
        n_extreme_low = (df_per_kwh[per_kwh_col] < y_min).sum()
        
        print(f"\n{col}:")
        print(f"  Minimum: ${min_val:.4f}/kWh")
        print(f"  Maximum: ${max_val:.4f}/kWh")
        print(f"  95th percentile: ${p95:.4f}/kWh")
        print(f"  98th percentile: ${p98:.4f}/kWh")
        print(f"  99th percentile: ${p99:.4f}/kWh")
        print(f"  Buildings beyond upper y-axis ({y_max:.4f}): {n_extreme_high} ({n_extreme_high/len(df_per_kwh)*100:.1f}%)")
        if n_extreme_low > 0:
            print(f"  Buildings beyond lower y-axis ({y_min:.4f}): {n_extreme_low} ({n_extreme_low/len(df_per_kwh)*100:.1f}%)")
    
    # Free memory after analysis
    del df_per_kwh
else:
    print("Column 'annual_kwh' not found; cannot compute cost per kWh.")
```

### Tariff deltas: who wins or loses relative to flat?

Here we compute, for each household, the **difference in annual bills** under TOU and Dynamic tariffs relative to a Flat tariff (positive values = higher bills, negative values = savings).  
We examine:
- **Overall distribution** of bill differences across all buildings (histograms)
- **Winner/loser analysis**: How many buildings pay more vs. less, and by how much?
- **Distributional equity**: How bill changes vary across Federal Poverty Level categories

This analysis helps identify whether alternative tariffs create systematic winners and losers, and whether impacts are equitably distributed across income groups.

```python
# Compare annual bill differences (TOU vs Flat, Dynamic vs Flat)

if bills_with_meta is not None:
    # Ensure the needed columns exist
    needed_cols = ["annual_cost_flat", "annual_cost_tou", "annual_cost_dynamic"]

    missing = [c for c in needed_cols if c not in bills_with_meta.columns]
    if missing:
        print("Missing bill columns, cannot compute deltas:", missing)
    else:
        # Compute deltas directly on bills_with_meta (memory-optimized: no copy)
        # Only create delta columns if they don't exist yet
        if "delta_tou_minus_flat" not in bills_with_meta.columns:
            bills_with_meta["delta_tou_minus_flat"] = (
                bills_with_meta["annual_cost_tou"] - bills_with_meta["annual_cost_flat"]
            )
        
        if "delta_dynamic_minus_flat" not in bills_with_meta.columns:
            bills_with_meta["delta_dynamic_minus_flat"] = (
                bills_with_meta["annual_cost_dynamic"] - bills_with_meta["annual_cost_flat"]
            )
        
        if "delta_dynamic_minus_tou" not in bills_with_meta.columns:
            bills_with_meta["delta_dynamic_minus_tou"] = (
                bills_with_meta["annual_cost_dynamic"] - bills_with_meta["annual_cost_tou"]
            )

        # ========== COMPARISON 1: TOU vs Flat ==========
        
        # --- Graph: Overall histogram for TOU - Flat ---
        plt.figure(figsize=(8, 4))
        bills_with_meta["delta_tou_minus_flat"].hist(bins=40)
        plt.axvline(0, color="k", linestyle="--")
        plt.xlabel("Annual bill difference: TOU - Flat ($)")
        plt.ylabel("Count")
        plt.title("Who wins/loses under TOU vs Flat?")
        plt.tight_layout()
        plt.show()

        # --- Analysis: TOU vs Flat ---
        tou_losers = bills_with_meta[bills_with_meta["delta_tou_minus_flat"] > 0].copy()
        tou_winners = bills_with_meta[bills_with_meta["delta_tou_minus_flat"] <= 0].copy()
        
        print("\n" + "="*70)
        print("ANALYSIS: Who loses under TOU vs Flat?")
        print("="*70)
        print(f"Buildings that LOSE (pay more): {len(tou_losers)} ({len(tou_losers)/len(bills_with_meta)*100:.1f}%)")
        print(f"Buildings that WIN (pay less): {len(tou_winners)} ({len(tou_winners)/len(bills_with_meta)*100:.1f}%)")
        
        if len(tou_losers) > 0:
            print(f"\nLosers - Annual kWh consumption:")
            print(f"  Mean: {tou_losers['annual_kwh'].mean():.0f} kWh")
            print(f"  Median: {tou_losers['annual_kwh'].median():.0f} kWh")
            print(f"\nLosers - Bill increase (TOU - Flat):")
            print(f"  Mean: ${tou_losers['delta_tou_minus_flat'].mean():.2f}")
            print(f"  Median: ${tou_losers['delta_tou_minus_flat'].median():.2f}")
            print(f"  Max: ${tou_losers['delta_tou_minus_flat'].max():.2f}")
            
        if len(tou_winners) > 0:
            print(f"\nWinners - Annual kWh consumption:")
            print(f"  Mean: {tou_winners['annual_kwh'].mean():.0f} kWh")
            print(f"  Median: {tou_winners['annual_kwh'].median():.0f} kWh")
            print(f"\nWinners - Bill savings (Flat - TOU):")
            print(f"  Mean: ${-tou_winners['delta_tou_minus_flat'].mean():.2f}")
            print(f"  Median: ${-tou_winners['delta_tou_minus_flat'].median():.2f}")

        # ========== COMPARISON 2: Dynamic vs Flat ==========
        
        # --- Graph: Overall histogram for Dynamic - Flat ---
        plt.figure(figsize=(8, 4))
        bills_with_meta["delta_dynamic_minus_flat"].hist(bins=40)
        plt.axvline(0, color="k", linestyle="--")
        plt.xlabel("Annual bill difference: Dynamic - Flat ($)")
        plt.ylabel("Count")
        plt.title("Who wins/loses under Dynamic vs Flat?")
        plt.tight_layout()
        plt.show()

        # --- Analysis: Dynamic vs Flat ---
        dynamic_losers = bills_with_meta[bills_with_meta["delta_dynamic_minus_flat"] > 0].copy()
        dynamic_winners = bills_with_meta[bills_with_meta["delta_dynamic_minus_flat"] <= 0].copy()
        
        print("\n" + "="*70)
        print("ANALYSIS: Who loses under Dynamic vs Flat?")
        print("="*70)
        print(f"Buildings that LOSE (pay more): {len(dynamic_losers)} ({len(dynamic_losers)/len(bills_with_meta)*100:.1f}%)")
        print(f"Buildings that WIN (pay less): {len(dynamic_winners)} ({len(dynamic_winners)/len(bills_with_meta)*100:.1f}%)")
        
        if len(dynamic_losers) > 0:
            print(f"\nLosers - Annual kWh consumption:")
            print(f"  Mean: {dynamic_losers['annual_kwh'].mean():.0f} kWh")
            print(f"  Median: {dynamic_losers['annual_kwh'].median():.0f} kWh")
            print(f"\nLosers - Bill increase (Dynamic - Flat):")
            print(f"  Mean: ${dynamic_losers['delta_dynamic_minus_flat'].mean():.2f}")
            print(f"  Median: ${dynamic_losers['delta_dynamic_minus_flat'].median():.2f}")
            print(f"  Max: ${dynamic_losers['delta_dynamic_minus_flat'].max():.2f}")
            
        if len(dynamic_winners) > 0:
            print(f"\nWinners - Annual kWh consumption:")
            print(f"  Mean: {dynamic_winners['annual_kwh'].mean():.0f} kWh")
            print(f"  Median: {dynamic_winners['annual_kwh'].median():.0f} kWh")
            print(f"\nWinners - Bill savings (Flat - Dynamic):")
            print(f"  Mean: ${-dynamic_winners['delta_dynamic_minus_flat'].mean():.2f}")
            print(f"  Median: ${-dynamic_winners['delta_dynamic_minus_flat'].median():.2f}")

        # ========== COMPARISON 3: Dynamic vs TOU ==========
        
        # --- Graph: Histogram for Dynamic - TOU ---
        plt.figure(figsize=(8, 4))
        bills_with_meta["delta_dynamic_minus_tou"].hist(bins=40)
        plt.axvline(0, color="k", linestyle="--")
        plt.xlabel("Annual bill difference: Dynamic - TOU ($)")
        plt.ylabel("Count")
        plt.title("Who wins/loses under Dynamic vs TOU?")
        plt.tight_layout()
        plt.show()
        
        # --- Analysis: Dynamic vs TOU ---
        dynamic_tou_losers = bills_with_meta[bills_with_meta["delta_dynamic_minus_tou"] > 0].copy()
        dynamic_tou_winners = bills_with_meta[bills_with_meta["delta_dynamic_minus_tou"] <= 0].copy()
        
        print("\n" + "="*70)
        print("ANALYSIS: Who loses under Dynamic vs TOU?")
        print("="*70)
        print(f"Buildings that LOSE (pay more): {len(dynamic_tou_losers)} ({len(dynamic_tou_losers)/len(bills_with_meta)*100:.1f}%)")
        print(f"Buildings that WIN (pay less): {len(dynamic_tou_winners)} ({len(dynamic_tou_winners)/len(bills_with_meta)*100:.1f}%)")
        
        if len(dynamic_tou_losers) > 0:
            print(f"\nLosers - Annual kWh consumption:")
            print(f"  Mean: {dynamic_tou_losers['annual_kwh'].mean():.0f} kWh")
            print(f"  Median: {dynamic_tou_losers['annual_kwh'].median():.0f} kWh")
            print(f"\nLosers - Bill increase (Dynamic - TOU):")
            print(f"  Mean: ${dynamic_tou_losers['delta_dynamic_minus_tou'].mean():.2f}")
            print(f"  Median: ${dynamic_tou_losers['delta_dynamic_minus_tou'].median():.2f}")
            print(f"  Max: ${dynamic_tou_losers['delta_dynamic_minus_tou'].max():.2f}")
            
        if len(dynamic_tou_winners) > 0:
            print(f"\nWinners - Annual kWh consumption:")
            print(f"  Mean: {dynamic_tou_winners['annual_kwh'].mean():.0f} kWh")
            print(f"  Median: {dynamic_tou_winners['annual_kwh'].median():.0f} kWh")
            print(f"\nWinners - Bill savings (TOU - Dynamic):")
            print(f"  Mean: ${-dynamic_tou_winners['delta_dynamic_minus_tou'].mean():.2f}")
            print(f"  Median: ${-dynamic_tou_winners['delta_dynamic_minus_tou'].median():.2f}")

        # ========== INCOME ANALYSIS ==========

        if "in.federal_poverty_level" in bills_with_meta.columns:
            # Filter out "Not Available" values for cleaner plots
            df_filtered = bills_with_meta[
                (bills_with_meta["in.federal_poverty_level"].notna()) &
                (bills_with_meta["in.federal_poverty_level"] != "Not Available")
            ]
            
            if not df_filtered.empty:
                # TOU vs Flat by income
                plt.figure(figsize=(6, 4))
                df_filtered.boxplot(
                    column="delta_tou_minus_flat",
                    by="in.federal_poverty_level",
                    grid=False,
                )
                plt.axhline(0, color="k", linestyle="--")
                plt.title("TOU - Flat bill difference by federal poverty level")
                plt.suptitle("")
                plt.xlabel("Federal poverty level")
                plt.ylabel("TOU - Flat ($)")
                plt.tight_layout()
                plt.show()
                
                print("\n" + "="*70)
                print("TOU - Flat bill difference by Federal Poverty Level")
                print("="*70)
                tou_stats = df_filtered.groupby("in.federal_poverty_level")["delta_tou_minus_flat"].agg([
                    'count', 'mean', 'median', 'std', 'min', 'max'
                ])
                tou_stats.columns = ['Count', 'Mean ($)', 'Median ($)', 'Std ($)', 'Min ($)', 'Max ($)']
                display(tou_stats)
            
                # Dynamic vs Flat by income
                plt.figure(figsize=(6, 4))
                df_filtered.boxplot(
                    column="delta_dynamic_minus_flat",
                    by="in.federal_poverty_level",
                    grid=False,
                )
                plt.axhline(0, color="k", linestyle="--")
                plt.title("Dynamic - Flat bill difference by federal poverty level")
                plt.suptitle("")
                plt.xlabel("Federal poverty level")
                plt.ylabel("Dynamic - Flat ($)")
                plt.tight_layout()
                plt.show()
                
                print("\n" + "="*70)
                print("Dynamic - Flat bill difference by Federal Poverty Level")
                print("="*70)
                dynamic_stats = df_filtered.groupby("in.federal_poverty_level")["delta_dynamic_minus_flat"].agg([
                    'count', 'mean', 'median', 'std', 'min', 'max'
                ])
                dynamic_stats.columns = ['Count', 'Mean ($)', 'Median ($)', 'Std ($)', 'Min ($)', 'Max ($)']
                display(dynamic_stats)
```

### Energy Bill Burden Analysis: Percentage of Income Spent on Utilities (PIU)

Energy bill burden, measured as the **Percentage of Income spent on Utilities (PIU)**, is a critical equity metric that reveals whether energy costs are affordable relative to household income.  

In this section, we compute PIU for each building under each tariff structure (flat, TOU, dynamic) by dividing annual electricity bills by representative household income.  

We examine:
- **PIU distribution** across income levels: How does the share of income spent on electricity vary between low-income and high-income households?
- **Tariff impact on burden**: Do different tariff structures (flat, TOU, dynamic) shift the burden differently across income groups?
- **Equity implications**: Are low-income households disproportionately burdened by electricity costs?

**Key metric**: PIU (%) = (Annual Electricity Bill / Annual Household Income) × 100

High PIU values (typically >6% for energy, >3% for electricity alone) indicate energy insecurity and potential hardship.

```python
# Compute PIU (Percentage of Income spent on Utilities) for each tariff

if bills_with_meta is not None:
    # Check if we have both income and bill data
    required_cols = ["in.representative_income", "annual_cost_flat", "annual_cost_tou", "annual_cost_dynamic"]
    missing = [c for c in required_cols if c not in bills_with_meta.columns]
    
    if missing:
        print(f"Missing required columns: {missing}")
    else:
        # Memory-efficient: Compute PIU directly on bills_with_meta without intermediate copies
        # Create new columns directly on the source DataFrame
        piu_cols = ["piu_flat", "piu_tou", "piu_dynamic"]
        
        # Vectorized computation of PIU for all tariffs
        bills_with_meta["piu_flat"] = (bills_with_meta["annual_cost_flat"] / bills_with_meta["in.representative_income"]) * 100
        bills_with_meta["piu_tou"] = (bills_with_meta["annual_cost_tou"] / bills_with_meta["in.representative_income"]) * 100
        bills_with_meta["piu_dynamic"] = (bills_with_meta["annual_cost_dynamic"] / bills_with_meta["in.representative_income"]) * 100
        
        # Filter to valid data: income > $1000 and non-negative PIU values
        # Using a combined filter avoids multiple intermediate DataFrames
        df_piu = bills_with_meta[
            (bills_with_meta["in.representative_income"].notna()) &
            (bills_with_meta["in.representative_income"] > 1000) &
            (bills_with_meta["piu_flat"] >= 0) &
            (bills_with_meta["piu_tou"] >= 0) &
            (bills_with_meta["piu_dynamic"] >= 0)
        ]
        
        # Report filtering results
        excluded = len(bills_with_meta) - len(df_piu)
        print(f"Computing PIU for {len(df_piu)} buildings with valid income data")
        print(f"(Excluded {excluded} buildings with missing/invalid income or negative PIU values)")
        
        # Summary statistics for PIU
        print("\n=== PIU Summary Statistics (%) ===")
        piu_summary = df_piu[piu_cols].describe()
        display(piu_summary)
        
        # Count buildings with high energy burden (>6% for total energy, using >3% for electricity only)
        print("\n=== Energy Burden Thresholds ===")
        for col in piu_cols:
            tariff_name = col.replace("piu_", "").upper()
            high_burden = (df_piu[col] > 3.0).sum()
            pct = (high_burden / len(df_piu)) * 100
            print(f"{tariff_name}: {high_burden} buildings ({pct:.1f}%) with PIU > 3%")
            
            very_high_burden = (df_piu[col] > 6.0).sum()
            pct_vh = (very_high_burden / len(df_piu)) * 100
            print(f"{tariff_name}: {very_high_burden} buildings ({pct_vh:.1f}%) with PIU > 6%")
        
        # Store df_piu for use in subsequent cells
        print(f"\ndf_piu created with shape: {df_piu.shape}")
else:
    print("bills_with_meta is None; cannot compute PIU")
```

```python
# PIU by Federal Poverty Level: Boxplots showing burden distribution

if 'df_piu' in locals() and "in.federal_poverty_level" in df_piu.columns:
    # Filter out "Not Available" poverty levels
    df_piu_poverty = df_piu[
        (df_piu["in.federal_poverty_level"].notna()) &
        (df_piu["in.federal_poverty_level"] != "Not Available")
    ]
    
    if df_piu_poverty.empty:
        print("No valid poverty level data for PIU analysis")
    else:
        print(f"Analyzing PIU for {len(df_piu_poverty)} buildings across poverty levels")
        
        piu_cols = ["piu_flat", "piu_tou", "piu_dynamic"]
        tariff_labels = {"piu_flat": "Flat", "piu_tou": "TOU", "piu_dynamic": "Dynamic"}
        
        # Create boxplots for each tariff
        for col in piu_cols:
            plt.figure(figsize=(8, 5))
            df_piu_poverty.boxplot(
                column=col,
                by="in.federal_poverty_level",
                grid=False,
            )
            plt.axhline(3.0, color="orange", linestyle="--", linewidth=1.5, label="3% threshold")
            plt.axhline(6.0, color="red", linestyle="--", linewidth=1.5, label="6% threshold")
            plt.title(f"Energy Bill Burden ({tariff_labels[col]} Tariff) by Income Level")
            plt.suptitle("")
            plt.xlabel("Federal Poverty Level")
            plt.ylabel("PIU: % of Income Spent on Electricity")
            plt.legend()
            # Set y-axis to 30% to show upper tails of low-income boxes
            plt.ylim(0, 200)
            plt.tight_layout()
            plt.show()
        
        # Summary statistics table
        print("\n=== PIU by Federal Poverty Level: Summary Statistics ===")
        piu_by_poverty = df_piu_poverty.groupby("in.federal_poverty_level")[piu_cols].agg([
            'count', 'mean', 'median', 'std', 'min', 'max'
        ])
        piu_by_poverty.columns = [f"{col}_{stat}" for col, stat in piu_by_poverty.columns]
        display(piu_by_poverty)
        
        # Count high-burden buildings by poverty level
        print("\n=== High Energy Burden (>6% PIU) by Poverty Level ===")
        for col in piu_cols:
            tariff_name = tariff_labels[col]
            high_burden_by_poverty = df_piu_poverty[df_piu_poverty[col] > 6.0].groupby("in.federal_poverty_level").size()
            total_by_poverty = df_piu_poverty.groupby("in.federal_poverty_level").size()
            pct_by_poverty = (high_burden_by_poverty / total_by_poverty * 100).fillna(0)
            
            print(f"\n{tariff_name} Tariff:")
            for poverty_level in df_piu_poverty["in.federal_poverty_level"].unique():
                count = high_burden_by_poverty.get(poverty_level, 0)
                total = total_by_poverty.get(poverty_level, 0)
                pct = pct_by_poverty.get(poverty_level, 0)
                print(f"  {poverty_level}: {count}/{total} buildings ({pct:.1f}%)")
else:
    print("df_piu not available or missing poverty level data")
```

### Progressivity Index:

Annual Cost= α + β⋅Income

A larger positive slope means costs rise more steeply with income. A smaller slope means the rate structure is less progressive, because the increase in cost across income groups is flatter. Dynamic pricing compresses annual cost differences across income levels the most, making it the least progressive structure, while flat pricing maintains the steepest increase in cost with income and is therefore the most progressive Lower-income groups see smaller discounts from TOU and dynamic. Higher-income groups see larger discounts, flattening the cost–income gradient.

```python
# Progressivity Index Analysis
# A tariff is "progressive" if costs rise more steeply with income
# Slope calculation: Annual Cost = α + β·Income
# Larger positive β = more progressive (costs scale with income)

# Memory-efficient approach: work directly with bills_with_meta, no copies
if 'bills_with_meta' in locals() and 'in.federal_poverty_level' in bills_with_meta.columns:
    
    # Define bill columns to analyze
    bill_cols = ['annual_cost_flat', 'annual_cost_tou', 'annual_cost_dynamic']
    
    # Create filter mask (memory efficient - boolean array)
    valid_mask = (
        (bills_with_meta['in.federal_poverty_level'].notna()) &
        (bills_with_meta['in.federal_poverty_level'] != 'Not Available') &
        (bills_with_meta[bill_cols[0]].notna()) &
        (bills_with_meta[bill_cols[1]].notna()) &
        (bills_with_meta[bill_cols[2]].notna())
    )
    
    n_valid = valid_mask.sum()
    print(f"Analyzing {n_valid:,} buildings with complete poverty level and bill data")
    
    if n_valid == 0:
        print("No valid data for progressivity analysis")
    else:
        # Define the ordering for Federal Poverty Level categories
        fpl_order = ['0-100%', '100-150%', '150-200%', '200-300%', '300-400%', '400%+']
        
        # Get unique FPL values from filtered data (memory efficient)
        available_fpl = bills_with_meta.loc[valid_mask, 'in.federal_poverty_level'].unique()
        available_fpl = [fpl for fpl in fpl_order if fpl in available_fpl]
        
        if len(available_fpl) == 0:
            print("No standard FPL categories found. Available values:")
            print(bills_with_meta.loc[valid_mask, 'in.federal_poverty_level'].unique())
        else:
            # Create second filter for standard FPL categories only
            fpl_mask = valid_mask & bills_with_meta['in.federal_poverty_level'].isin(available_fpl)
            
            print(f"\nFederal Poverty Level distribution:")
            print(bills_with_meta.loc[fpl_mask, 'in.federal_poverty_level'].value_counts().reindex(available_fpl))
            
            # Map FPL categories to numeric values for regression
            fpl_to_num = {fpl: i+1 for i, fpl in enumerate(available_fpl)}
            
            # Calculate grouped statistics (only compute what we need)
            df_inc = bills_with_meta.loc[fpl_mask].groupby('in.federal_poverty_level', observed=True)[bill_cols].mean()
            
            # Add numeric FPL values and reindex to ensure correct order
            df_inc = df_inc.reindex(available_fpl).reset_index()
            df_inc['fpl_num'] = df_inc['in.federal_poverty_level'].map(fpl_to_num)
            
            # Calculate progressivity slopes for each tariff
            print("\n=== Progressivity Analysis ===")
            print("Regression: Annual Cost = α + β·(FPL Group)")
            print("Larger β indicates more progressive tariff (costs rise faster with income)\n")
            
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
            
            # Plot 1: Average cost by income group with regression lines
            x = df_inc['fpl_num'].values
            
            tariff_colors = {
                'annual_cost_flat': '#1f77b4',
                'annual_cost_tou': '#ff7f0e', 
                'annual_cost_dynamic': '#2ca02c'
            }
            tariff_names = {
                'annual_cost_flat': 'Flat Rate',
                'annual_cost_tou': 'Time-of-Use',
                'annual_cost_dynamic': 'Dynamic Pricing'
            }
            
            slopes = {}
            intercepts = {}
            
            for col in bill_cols:
                y = df_inc[col].values
                
                # Calculate regression line
                slope, intercept = np.polyfit(x, y, 1)
                slopes[col] = slope
                intercepts[col] = intercept
                
                # Plot points and line
                ax1.scatter(x, y, alpha=0.7, s=100, color=tariff_colors[col], 
                           label=f"{tariff_names[col]}")
                
                x_line = np.linspace(x.min(), x.max(), 100)
                y_line = slope * x_line + intercept
                ax1.plot(x_line, y_line, '--', color=tariff_colors[col], linewidth=2)
            
            ax1.set_xticks(list(fpl_to_num.values()))
            ax1.set_xticklabels(list(fpl_to_num.keys()), rotation=15, ha='right')
            ax1.set_xlabel("Federal Poverty Level", fontsize=12)
            ax1.set_ylabel("Annual Cost ($)", fontsize=12)
            ax1.set_title("Average Annual Cost by Income Group", fontsize=14, fontweight='bold')
            ax1.legend(fontsize=10)
            ax1.grid(True, alpha=0.3)
            
            # Plot 2: Progressivity index comparison (slope values)
            slopes_sorted = sorted(slopes.items(), key=lambda x: x[1], reverse=True)
            tariff_labels = [tariff_names[t] for t, _ in slopes_sorted]
            slope_values = [s for _, s in slopes_sorted]
            colors = [tariff_colors[t] for t, _ in slopes_sorted]
            
            bars = ax2.barh(tariff_labels, slope_values, color=colors, alpha=0.7)
            ax2.set_xlabel("Progressivity Index (β - Slope)", fontsize=12)
            ax2.set_title("Tariff Progressivity Comparison", fontsize=14, fontweight='bold')
            ax2.grid(True, alpha=0.3, axis='x')
            
            # Add value labels on bars
            for bar, val in zip(bars, slope_values):
                ax2.text(val + 5, bar.get_y() + bar.get_height()/2, 
                        f'β = {val:.1f}', va='center', fontsize=11, fontweight='bold')
            
            plt.tight_layout()
            plt.show()
            
            # Print regression results
            print("\nRegression Results:")
            print("-" * 60)
            for col in bill_cols:
                print(f"{tariff_names[col]:20s}: Cost = {intercepts[col]:6.1f} + {slopes[col]:5.1f} × FPL Group")
                print(f"{'':20s}  Slope (β) = {slopes[col]:.1f}")
            
            # Interpretation
            print("\n" + "=" * 60)
            print("INTERPRETATION:")
            print("-" * 60)
            most_progressive = max(slopes, key=slopes.get)
            least_progressive = min(slopes, key=slopes.get)
            print(f"Most Progressive:  {tariff_names[most_progressive]} (β = {slopes[most_progressive]:.1f})")
            print(f"Least Progressive: {tariff_names[least_progressive]} (β = {slopes[least_progressive]:.1f})")
            print("\nA higher slope means costs rise more steeply with income,")
            print("indicating the tariff structure better reflects ability to pay.")
            print("\nDynamic pricing compresses cost differences across income levels,")
            print("making it less progressive because lower-income groups see smaller")
            print("discounts while higher-income groups see larger discounts.")
        
else:
    print("bills_with_meta not available or missing required columns")
```

### Geographical Context for Massachusetts Energy Analysis

To understand the spatial distribution of energy consumption and burden across Massachusetts, we analyze regional patterns in:

1. **Average Energy Consumption (kWh)**: Mean annual electricity use per building by region
2. **Total Energy Consumption (GWh)**: Aggregate electricity consumption across all buildings in each region
3. **Average PIU (%)**: Mean percentage of income spent on utilities by region

This regional breakdown helps identify geographic disparities in energy use and affordability, revealing whether certain areas face disproportionate energy burdens due to housing stock characteristics, climate conditions, or socioeconomic factors.

```python
# Regional energy consumption and burden analysis for Massachusetts

if bills_with_meta is not None:
    # Check for county or region column
    region_col = None
    for col in ["in.county_name", "in.ahs_region", "in.state"]:
        if col in bills_with_meta.columns:
            region_col = col
            break
    
    if region_col is None:
        print("No regional identifier column found in metadata")
    else:
        # Filter for valid data (non-null regions and PIU values)
        # First, ensure we have PIU columns computed
        if "piu_flat" not in bills_with_meta.columns:
            # Compute PIU if not already present
            if "in.representative_income" in bills_with_meta.columns:
                bills_with_meta["piu_flat"] = (bills_with_meta["annual_cost_flat"] / bills_with_meta["in.representative_income"]) * 100
                bills_with_meta["piu_tou"] = (bills_with_meta["annual_cost_tou"] / bills_with_meta["in.representative_income"]) * 100
                bills_with_meta["piu_dynamic"] = (bills_with_meta["annual_cost_dynamic"] / bills_with_meta["in.representative_income"]) * 100
        
        # Filter for valid region and PIU data
        df_regional = bills_with_meta[
            (bills_with_meta[region_col].notna()) &
            (bills_with_meta["annual_kwh"].notna())
        ].copy()
        
        if df_regional.empty:
            print(f"No valid regional data found in {region_col}")
        else:
            # Compute regional statistics
            regional_stats = df_regional.groupby(region_col).agg({
                'annual_kwh': ['mean', 'sum', 'count'],
                'piu_flat': 'mean',
                'piu_tou': 'mean',
                'piu_dynamic': 'mean'
            }).round(2)
            
            # Flatten column names
            regional_stats.columns = ['_'.join(col).strip() for col in regional_stats.columns.values]
            
            # Rename for clarity
            regional_stats = regional_stats.rename(columns={
                'annual_kwh_mean': 'Avg Energy (kWh)',
                'annual_kwh_sum': 'Total Energy (kWh)',
                'annual_kwh_count': 'Building Count',
                'piu_flat_mean': 'Avg PIU Flat (%)',
                'piu_tou_mean': 'Avg PIU TOU (%)',
                'piu_dynamic_mean': 'Avg PIU Dynamic (%)'
            })
            
            # Convert total energy to GWh (1 GWh = 1,000,000 kWh)
            regional_stats['Total Energy (GWh)'] = (regional_stats['Total Energy (kWh)'] / 1_000_000).round(3)
            
            # Reorder columns for better presentation
            regional_stats = regional_stats[[
                'Building Count',
                'Avg Energy (kWh)',
                'Total Energy (GWh)',
                'Avg PIU Flat (%)',
                'Avg PIU TOU (%)',
                'Avg PIU Dynamic (%)'
            ]]
            
            # Sort by total energy consumption (descending)
            regional_stats = regional_stats.sort_values('Total Energy (GWh)', ascending=False)
            
            print(f"Regional Energy Consumption and Burden Analysis ({region_col})")
            
            display(regional_stats)
            
            # Summary statistics
            print(f"\n{'='*80}")
            print("Summary Statistics Across All Regions")
            print(f"{'='*80}")
            print(f"Total buildings analyzed: {regional_stats['Building Count'].sum()}")
            print(f"Total MA energy consumption: {regional_stats['Total Energy (GWh)'].sum():.3f} GWh")
            print(f"Overall average consumption: {df_regional['annual_kwh'].mean():.0f} kWh/building")
            print(f"Overall average PIU (Flat): {df_regional['piu_flat'].mean():.2f}%")
            print(f"Overall average PIU (TOU): {df_regional['piu_tou'].mean():.2f}%")
            print(f"Overall average PIU (Dynamic): {df_regional['piu_dynamic'].mean():.2f}%")
else:
    print("bills_with_meta is None; cannot perform regional analysis")
```

```python
# Load Massachusetts county boundaries
ma_counties = load_ma_counties(verbose=False)

# Prepare regional stats for merging
regional_stats_for_merge = regional_stats.reset_index()
regional_stats_for_merge = regional_stats_for_merge.rename(columns={'in.county_name': 'County'})
regional_stats_for_merge['County_Clean'] = regional_stats_for_merge['County'].str.replace(' County', '', regex=False)

# Merge geographic boundaries with energy statistics
ma_counties_energy = ma_counties.merge(
    regional_stats_for_merge,
    left_on='NAME',
    right_on='County_Clean',
    how='left'
)

# Count buildings with county data
buildings_with_county = bills_with_meta['in.county_name'].notna().sum()
total_buildings = len(bills_with_meta)
counties_with_data = ma_counties_energy['Avg Energy (kWh)'].notna().sum()
total_counties = len(ma_counties_energy)

print(f"Buildings with county data: {buildings_with_county:,}/{total_buildings:,}")
print(f"Counties with energy data: {counties_with_data}/{total_counties}")
print()

# 1. Average Energy Consumption (kWh)
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
ma_counties_energy.plot(
    column='Avg Energy (kWh)',
    cmap='YlOrRd',
    legend=True,
    ax=ax,
    edgecolor='black',
    linewidth=0.5,
    legend_kwds={'label': 'Avg kWh per Building', 'shrink': 0.8},
    missing_kwds={'color': 'lightgrey'}
)
ax.set_title('Average Energy Consumption per Building by County', fontsize=14, fontweight='bold')
ax.axis('off')

# Add county labels
for idx, row in ma_counties_energy.iterrows():
    centroid = row.geometry.centroid
    ax.text(centroid.x, centroid.y, row['NAME'], 
            fontsize=8, ha='center', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

# 2. Total Energy Consumption (GWh)
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
ma_counties_energy.plot(
    column='Total Energy (GWh)',
    cmap='Greens',
    legend=True,
    ax=ax,
    edgecolor='black',
    linewidth=0.5,
    legend_kwds={'label': 'Total GWh', 'shrink': 0.8},
    missing_kwds={'color': 'lightgrey'}
)
ax.set_title('Total Energy Consumption (All Buildings) by County', fontsize=14, fontweight='bold')
ax.axis('off')

for idx, row in ma_counties_energy.iterrows():
    centroid = row.geometry.centroid
    ax.text(centroid.x, centroid.y, row['NAME'], 
            fontsize=8, ha='center', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

# 3. Average PIU - Flat Rate
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
ma_counties_energy.plot(
    column='Avg PIU Flat (%)',
    cmap='RdYlBu_r',
    legend=True,
    ax=ax,
    edgecolor='black',
    linewidth=0.5,
    legend_kwds={'label': 'PIU (%)', 'shrink': 0.8},
    missing_kwds={'color': 'lightgrey'}
)
ax.set_title('Average Percent Income for Utilities - Flat Rate by County', fontsize=14, fontweight='bold')
ax.axis('off')

for idx, row in ma_counties_energy.iterrows():
    centroid = row.geometry.centroid
    ax.text(centroid.x, centroid.y, row['NAME'], 
            fontsize=8, ha='center', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

# 4. Average PIU - Dynamic Pricing
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
ma_counties_energy.plot(
    column='Avg PIU Dynamic (%)',
    cmap='RdYlBu_r',
    legend=True,
    ax=ax,
    edgecolor='black',
    linewidth=0.5,
    legend_kwds={'label': 'PIU (%)', 'shrink': 0.8},
    missing_kwds={'color': 'lightgrey'}
)
ax.set_title('Average Percent Income for Utilities - Dynamic Pricing by County', fontsize=14, fontweight='bold')
ax.axis('off')

for idx, row in ma_counties_energy.iterrows():
    centroid = row.geometry.centroid
    ax.text(centroid.x, centroid.y, row['NAME'], 
            fontsize=8, ha='center', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

# Display summary statistics for counties with data
print("\nCounty Energy Statistics Summary:")
print("=" * 80)
display(ma_counties_energy[['NAME', 'Building Count', 'Avg Energy (kWh)', 
                             'Total Energy (GWh)', 'Avg PIU Flat (%)', 
                             'Avg PIU Dynamic (%)']].sort_values('Total Energy (GWh)', ascending=False))
```

## Geographic Analysis: PUMA-Level Heatmaps

PUMA (Public Use Microdata Area) boundaries provide a coarser regional view than counties. Massachusetts has 52 PUMA regions. Let's visualize the same energy metrics at this geographic level.

```python
# Create PUMA-level statistics
puma_region_col = 'in.puma'

if puma_region_col in bills_with_meta.columns:
    # Group by PUMA and calculate statistics
    puma_stats = bills_with_meta.groupby(puma_region_col).agg({
        'bldg_id': 'count',
        'annual_kwh': 'mean'
    }).rename(columns={
        'bldg_id': 'Building Count',
        'annual_kwh': 'Avg Energy (kWh)'
    })
    
    # Calculate total energy in GWh
    total_energy_gwh = bills_with_meta.groupby(puma_region_col)['annual_kwh'].sum() / 1_000_000
    puma_stats['Total Energy (GWh)'] = total_energy_gwh
    
    # Calculate average PIU for each tariff
    piu_cols = ['piu_flat', 'piu_tou', 'piu_dynamic']
    existing_piu = [col for col in piu_cols if col in bills_with_meta.columns]
    
    for col in existing_piu:
        tariff_name = col.replace('piu_', '').replace('_', ' ').title()
        puma_stats[f'Avg PIU {tariff_name} (%)'] = bills_with_meta.groupby(puma_region_col)[col].mean()
    
    # Load Massachusetts PUMA boundaries
    ma_puma = load_ma_puma(verbose=False)
    
    # Prepare PUMA stats for merging
    puma_stats_for_merge = puma_stats.reset_index()
    
    # Extract last 5 digits from PUMA code in energy data (format: G25000100 -> 00100)
    puma_stats_for_merge['PUMA_Clean'] = puma_stats_for_merge[puma_region_col].str[-5:]
    
    # Merge geographic boundaries with energy statistics
    ma_puma_energy = ma_puma.merge(
        puma_stats_for_merge,
        left_on='PUMACE10',
        right_on='PUMA_Clean',
        how='left'
    )
    
    # Count buildings with PUMA data
    buildings_with_puma = bills_with_meta[puma_region_col].notna().sum()
    total_buildings = len(bills_with_meta)
    pumas_with_data = ma_puma_energy['Avg Energy (kWh)'].notna().sum()
    total_pumas = len(ma_puma_energy)
    
    print(f"Buildings with PUMA data: {buildings_with_puma:,}/{total_buildings:,}")
    print(f"PUMAs with energy data: {pumas_with_data}/{total_pumas}")
    print()
    
    # 1. Average Energy Consumption (kWh)
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    ma_puma_energy.plot(
        column='Avg Energy (kWh)',
        cmap='YlOrRd',
        legend=True,
        ax=ax,
        edgecolor='black',
        linewidth=0.5,
        legend_kwds={'label': 'Avg kWh per Building', 'shrink': 0.8},
        missing_kwds={'color': 'lightgrey'}
    )
    ax.set_title('Average Energy Consumption per Building by PUMA', fontsize=14, fontweight='bold')
    ax.axis('off')
    plt.tight_layout()
    plt.show()
    
    # 2. Total Energy Consumption (GWh)
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    ma_puma_energy.plot(
        column='Total Energy (GWh)',
        cmap='Greens',
        legend=True,
        ax=ax,
        edgecolor='black',
        linewidth=0.5,
        legend_kwds={'label': 'Total GWh', 'shrink': 0.8},
        missing_kwds={'color': 'lightgrey'}
    )
    ax.set_title('Total Energy Consumption (All Buildings) by PUMA', fontsize=14, fontweight='bold')
    ax.axis('off')
    plt.tight_layout()
    plt.show()
    
    # 3. Average PIU - Flat Rate
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    ma_puma_energy.plot(
        column='Avg PIU Flat (%)',
        cmap='RdYlBu_r',
        legend=True,
        ax=ax,
        edgecolor='black',
        linewidth=0.5,
        legend_kwds={'label': 'PIU (%)', 'shrink': 0.8},
        missing_kwds={'color': 'lightgrey'}
    )
    ax.set_title('Average Percent Income for Utilities - Flat Rate by PUMA', fontsize=14, fontweight='bold')
    ax.axis('off')
    plt.tight_layout()
    plt.show()
    
    # 4. Average PIU - Dynamic Pricing
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    ma_puma_energy.plot(
        column='Avg PIU Dynamic (%)',
        cmap='RdYlBu_r',
        legend=True,
        ax=ax,
        edgecolor='black',
        linewidth=0.5,
        legend_kwds={'label': 'PIU (%)', 'shrink': 0.8},
        missing_kwds={'color': 'lightgrey'}
    )
    ax.set_title('Average Percent Income for Utilities - Dynamic Pricing by PUMA', fontsize=14, fontweight='bold')
    ax.axis('off')
    plt.tight_layout()
    plt.show()
    
    # Display summary statistics
    print("\nPUMA Energy Statistics Summary (Top 10 by Total Energy):")
    print("=" * 80)
    display(ma_puma_energy[['PUMACE10', 'NAMELSAD10', 'Building Count', 
                             'Avg Energy (kWh)', 'Total Energy (GWh)', 
                             'Avg PIU Flat (%)', 'Avg PIU Dynamic (%)']].sort_values(
                                 'Total Energy (GWh)', ascending=False).head(10))
else:
    print(f"No {puma_region_col} column found in data")
```
